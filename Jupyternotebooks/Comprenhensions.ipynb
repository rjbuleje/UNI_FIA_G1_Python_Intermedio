 
### **Class about List and Dictionary comprenhensions**

Comprenhensions allow to create list and dictionary in one code line without using a for-loop.

Ways to write a list comprehension

```
list = [element for element in iterable if condition]
list = [element if condition else result for interable for iterable]
```
Ways to write a dictionay comprehension
```
dictionary = {keys:values for element in interable if condition}
```
