```python
def comprobar_primo(primo):
    list_divisores = [i for i in range(1, primo+1) if primo % i == 0]
    if len(list_divisores) == 2:
        return True
    else:
        return False
# arreglar
```

```python
def numero_sheldon(primo):
    if comprobar_primo(primo):
        lista_enumerada = enumerate([i for i in range(1,primo+1) if comprobar_primo(i) == True])
        orden_primo_original = [ i+1 for (i, j) in lista_enumerada if j == primo]
        cifras_primo = [int(i) for i in str(primo)]
        multiplicacion_cifras_primo = 1
        for i in cifras_primo:
            multiplicacion_cifras_primo *= i
        if multiplicacion_cifras_primo == orden_primo_original[0]:
            lista_enumerada = enumerate([i for i in range(1,primo+1) if comprobar_primo(i) == True])
            primo_inverso = int(str(primo)[::-1])
            if comprobar_primo(primo_inverso):
                orden_primo_inverso = [ i+1 for (i, j) in lista_enumerada if j == primo_inverso]
                orden_primo_inverso = orden_primo_inverso[0]
                orden_primo_original = orden_primo_original[0]
                comparacion = orden_primo_inverso == int(str(orden_primo_original)[::-1])
                print(comparacion) 
                if comparacion:
                    print('Sheldon')
                else:
                    print('error 1')
            else:
                 print('error 2')       
        else: 
            print('error 3')      
    else:
        print('error 4')
```
